```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


```

```python
# Read the dataset
df = pd.read_csv('assignment4.csv')
```

```python
df
```

```python
# Missing Values
missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
# Handle Missing Values
# Example: Drop rows with missing values in 'last_review' column
df = df.dropna(subset=['last_review'])
```

```python
missing_values
```

```python
df
```

```python
missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
# Correct Data Types
# Convert 'last_review' column to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df
```

```python
df.head()
```

```python
# Address Inconsistencies
# Example: Standardize neighborhood names to title case
df['neighbourhood'] = df['neighbourhood'].str.title()
```

```python
df.head()
```

```python
# Remove Duplicates
df = df.drop_duplicates()
```

```python
df.contain()
```

```python
df.describe()
```

```python
# Visualize distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize relationship between price and number of reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=df, color='green', alpha=0.5)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
# Visualize average price by room type
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='price', data=df, palette='viridis')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```
